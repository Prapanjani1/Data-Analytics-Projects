```python
from tkinter import *
from tkinter import messagebox
root=Tk()
root.title("Railway Reservation Portal")
root.geometry("300x250")
root.config(bg="lightblue")
label=Label(root,text="Enter Your Credentials",font=("Times",14))
label.grid(row=0,column=0, columnspan=2)

root.grid_rowconfigure(0, weight=1)
root.grid_rowconfigure(1, weight=1)
root.grid_rowconfigure(2, weight=1)

root.grid_columnconfigure(0, weight=1)
root.grid_columnconfigure(1, weight=1)

user=Label(root,text="Username",font=("Times",14))
user.grid(row=1,column=0)
name=Entry(root)
name.grid(row=1,column=1)
password=Label(root,text="Password",font=("Times",14))
password.grid(row=2,column=0)
passname=Entry(root,show="*")
passname.grid(row=2,column=1)

def journey_details():
    window=Tk()
    window.title("Check Train Availability")
    window.config(bg="#FFFFE0")
    window.geometry("650x350")
    label=Label(window,text="Enter Journey Details",font=('Times',25))
    label.grid(row=0,column=0,columnspan=2)
    source_label = Label(window, text="Source:")
    source_label.grid(row=2,column=1)
    options = ['Chennai','Madurai','Dharmapuri','Salem','Erode','Coimbatore']
    source_var=StringVar(window)
    source_menu=OptionMenu(window,source_var,*options)
    source_menu.grid(row=2,column=8)

    destination_label = Label(window, text="Destination:")
    destination_label.grid(row=3,column=1)
    options = ['Chennai','Madurai','Dharmapuri','Salem','Erode','Coimbatore']
    destination_var=StringVar(window)
    destination_menu=OptionMenu(window,destination_var,*options)
    destination_menu.grid(row=3,column=8)

    date_label = Label(window, text="Date of Travel (DD-MM-YYYY):")
    date_label.grid(row=4,column=1)
    date_entry = Entry(window)
    date_entry.grid(row=4,column=8)

    passengers_label = Label(window, text="Number of Passengers:")
    passengers_label.grid(row=5,column=1)
    passengers_entry = Entry(window)
    passengers_entry.grid(row=5,column=8)
    # adults_label = Label(window, text="Adults")
    # adults_label.grid(row=6,column=2)
    # adults_entry = Entry(window,width=5)
    # adults_entry.grid(row=6,column=3)
    # children_label = Label(window, text="Children")
    # children_label.grid(row=6,column=4)
    # children_entry = Entry(window,width=5)
    # children_entry.grid(row=6,column=5)



    def check_availability():
        passengers = int(passengers_entry.get())
        available_seats = 50
        if passengers<=available_seats:
            messagebox.showinfo("Successful"," Tickets are available ")
        else:
            messagebox.showerror("Error", "Tickets are not available")
        
    check=Button(window,text="Check Availability",command=check_availability)
    check.grid(row=6,column=2,pady=5)
    

    def book_ticket():
        date = date_entry.get()
        passengers = passengers_entry.get()
    
        if date and passengers:
            messagebox.showinfo("Wait","Redirected to fill passengers details")
            window.destroy()
            def passenger_details():
                window1=Tk()
                window1.title("Passenger Details")
                window1.config(bg="lightgreen")
                window1.geometry("700x700")
                label1=Label(window1,text="Enter Personal Details",font=("Arial",25))
                label1.grid(row=0,column=0,columnspan=2)
                window1.grid_rowconfigure(0, weight=1)
                window1.grid_rowconfigure(1, weight=1)
                window1.grid_rowconfigure(2, weight=1)
                window1.grid_rowconfigure(3, weight=1)
                window1.grid_rowconfigure(4, weight=1)
                window1.grid_rowconfigure(5, weight=1)
                window1.grid_rowconfigure(6, weight=1)
                window1.grid_rowconfigure(7, weight=1)
                window1.grid_rowconfigure(8, weight=1)
                window1.grid_rowconfigure(9, weight=1)
                window1.grid_columnconfigure(0, weight=1)
                window1.grid_columnconfigure(2, weight=1)
                window1.grid_columnconfigure(3, weight=1)
                window1.grid_columnconfigure(4, weight=1)
                name_label = Label(window1, text="Name:")
                name_label.grid(row=1,column=0)
                name_entry = Entry(window1)
                name_entry.grid(row=1,column=4)
                age_label = Label(window1, text="Age:")
                age_label.grid(row=2,column=0)
                age_entry = Entry(window1,width=5)
                age_entry.grid(row=2,column=4)
                gender_label=Label(window1,text="Gender:").grid(row=3,column=0)
                male=Radiobutton(window1,text='Male',value=1)
                male.grid(row=3,column=1)
                female=Radiobutton(window1,text='Female',value=2)
                female.grid(row=3,column=2)
                others=Radiobutton(window1,text='Others',value=3)
                others.grid(row=3,column=3)
                mobile_label=Label(window1,text="Mobile:",width=8).grid(row=4,column=0)
                code_var=StringVar(window1)
                mobile_menu=OptionMenu(window1,code_var,'+91','+1','+44',"+64")
                mobile_menu.grid(row=4,column=3)
                mobile_entry = Entry(window1)
                mobile_entry.grid(row=4,column=4)
                address_label=Label(window1,text="Address:").grid(row=5,column=0)
                address_text=Text(window1,wrap='word',width=20,height=5).grid(row=5,column=4)
                
                label2=Label(window1,text="Select Preference",font=("Arial",25))
                label2.grid(row=6,column=0,columnspan=2)
                food=Label(window1,text="Food").grid(row=7,column=0)
                breakfast=Checkbutton(window1,text="Breakfast").grid(row=7,column=1)
                lunch=Checkbutton(window1,text="Lunch").grid(row=7,column=2)
                dinner=Checkbutton(window1,text="Dinner").grid(row=7,column=3)
                category=Label(window1,text="Category").grid(row=8,column=1)
                veg=Checkbutton(window1,text="Vegetarian").grid(row=8,column=2)
                def on_select(event):
                    selected_indices = listbox.curselection()
                    selected_items = [listbox.get(idx) for idx in selected_indices]
                    selected.config(text="Selected: " + ", ".join(selected_items))
                listbox = Listbox(window1, selectmode=MULTIPLE,width=20,height=6)
                listbox.grid(row=8,column=3)
                for item in ["Sambar rice", "Veg Biriyani", "Meals", "Curd rice", "Chapathi kurma", "Tiffen"]:
                    listbox.insert(END, item)
                listbox.bind("<<ListboxSelect>>", on_select)
                selected=Label(window1,text='Selected:')
                selected.grid(row=8,column=4)
                non_veg=Checkbutton(window1,text="Non-Vegetarian").grid(row=9,column=2)
                def on_select1(event1):
                    selected_indices1=list.curselection()
                    selected_items1=[list.get(xy)for xy in selected_indices1]
                    select.config(text="Selected: " + ",".join(selected_items1))
                list=Listbox(window1,selectmode=MULTIPLE,width=20,height=6)
                list.grid(row=9,column=3)
                for items in ["Chicken rice","Non-Veg Meals","Fish Curry","Chicken Curry","Mutton Biriyani","Mutton Curry"]:
                    list.insert(END,items)
                list.bind("<<ListboxSelect>>",on_select1)
                select=Label(window1,text="Selected:")
                select.grid(row=9,column=4)
                berth=Label(window1,text="Berth").grid(row=10,column=0)
                def menu_command(option):
                    selected_label.config(text="Selected : " + option)
                menu_button = Menubutton(window1, text="Select Berth", relief="raised")
                menu = Menu(menu_button, tearoff=0)
                menu.add_command(label="Lower", command=lambda: menu_command("Lower"))
                menu.add_command(label="Upper", command=lambda: menu_command("Upper"))
                menu.add_command(label="Middle", command=lambda: menu_command("Middle"))
                menu_button.config(menu=menu)
                menu_button.grid(row=10,column=1)
                selected_label = Label(window1,text="Selected : ")
                selected_label.grid(row=10,column=4)
                coach_label=Label(window1,text="Coach").grid(row=11,column=0)
                def coach_command(option):
                    select_label.config(text="Selected: " + option)
                coach_button = Menubutton(window1, text="Select Coach", relief="raised")
                coach = Menu(coach_button, tearoff=0)
                coach.add_command(label="AC", command=lambda: coach_command("AC"))
                coach.add_command(label="Non AC", command=lambda: coach_command("Non AC"))
                coach.add_command(label="Any", command=lambda: coach_command("Any"))
                coach_button.config(menu=coach)
                coach_button.grid(row=11,column=1)
                select_label = Label(window1,text="Selected : ")
                select_label.grid(row=11,column=4)
                def make_payment():
                    messagebox.showinfo("Please Wait","You are redirected to make payment")
                    window1.destroy()
                    def payment_details():
                        payment_window=Tk()
                        payment_window.title("Payment Details")
                        payment_window.config(bg="lightgreen")
                        payment_window.geometry("600x400")
                        label1=Label(payment_window,text="WELCOME TO SELECT PAYMENT",font=("Arial",25))
                        label1.grid(row=0,column=0)
                        payment_window.grid_rowconfigure(0, weight=1)
                        payment_window.grid_rowconfigure(1, weight=1)
                        payment_window.grid_rowconfigure(2, weight=1)
                        payment_window.grid_rowconfigure(3, weight=1)
                        payment_window.grid_rowconfigure(4, weight=1)
                        payment_window.grid_rowconfigure(5, weight=1)
                        payment_window.grid_columnconfigure(0, weight=1)
                        payment_window.grid_columnconfigure(1, weight=1)
                        payment_window.grid_columnconfigure(1, weight=1)
                        label2=Label(payment_window,text="Select Payment Option",font=("Times",18))
                        label2.grid(row=1,column=0)
                        def offline():
                            label_offline=Label(payment_window,text="Enter Your Details")
                            label_offline.grid(row=2,column=0)
                            label_name=Label(payment_window,text="Name")
                            label_name.grid(row=3,column=0)
                            entry_name1=Entry(payment_window)
                            entry_name1.grid(row=3,column=1)
                            label_station=Label(payment_window,text="Station Name")
                            label_station.grid(row=4,column=0)
                            entry_station_name=Entry(payment_window)
                            entry_station_name.grid(row=4,column=1)
                            def on_click():
                                messagebox.showinfo("Successful",("Make sure to complete your payment before your journey"),("Your Tickets are booked"))
                            button_submit=Button(payment_window,text="Submit",command=on_click)
                            button_submit.grid(row=5,column=2)
                        button1=Button(payment_window,text="Offline",fg='red',command=offline)
                        button1.grid(row=1,column=1)
                        def online():
                            messagebox.showinfo("Please Wait","Redirected to make online payment")
                            payment_window.destroy()
                            def online_payment():
                                online_window=Tk()
                                online_window.title("Welcome to Online Payment Portal") 
                                online_window.config(bg="lightgreen")
                                online_window.geometry("600x350")
                                on_label=Label(online_window,text="Select Payment Mode",font=("Times",24))
                                on_label.grid(row=0,column=0)
                                online_window.grid_rowconfigure(0, weight=1)
                                online_window.grid_rowconfigure(1, weight=1)
                                online_window.grid_rowconfigure(2, weight=1)
                                online_window.grid_columnconfigure(0, weight=1)
                                online_window.grid_columnconfigure(1, weight=1)
                                online_window.grid_columnconfigure(2, weight=1)
                                online_window.grid_columnconfigure(3, weight=1)
                                on_button=Radiobutton(online_window,text="Credit Card",font=("Times",14),value=1)
                                on_button.grid(row=1,column=1)
                                on_button1=Radiobutton(online_window,text="Debit Card",font=("Times",14),value=2)
                                on_button1.grid(row=1,column=2)
                                on_button2=Radiobutton(online_window,text="UPI",font=("Times",14),value=3)
                                on_button2.grid(row=1,column=3)
                                def on_submit():
                                    messagebox.showinfo("Successful","Payment Successful")
                                    online_window.destroy()
                                button3=Button(online_window,text="SUBMIT",fg="red",command=on_submit)
                                button3.grid(row=2,column=2)
                            online_payment()
                        button2=Button(payment_window,text="Online",fg="red",command=online)
                        button2.grid(row=1,column=2)
                    payment_details()
                payment=Button(window1,text="Make Payment",command=make_payment)
                payment.grid(row=12,column=2)
                window1.mainloop()
            passenger_details()  
        else:
            messagebox.showerror("Error", "Please fill in all fields.")

    book_button = Button(window, text="Book Ticket", command=book_ticket)
    book_button.grid(row=7,column=2,pady=5)
    window.mainloop()
def check_credentials():
    username="user"
    password="pass"
    entered_username=name.get()
    entered_password=passname.get()
    if entered_username==username and entered_password==password:
        root.destroy()
        journey_details()
    else:
        error_label=Label(root,text="Invalid Username and/or Password",fg='red')
        error_label.grid(row=4,column=1)
submit=Button(root,text="Submit",command=check_credentials)
submit.grid(row=3,column=0)
def cancel_window():
    root.destroy()
cancel=Button(root,text="Cancel",command=cancel_window)
cancel.grid(row=3,column=1)
root.mainloop()
```
