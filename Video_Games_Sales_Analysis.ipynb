```python
# Calculating Confidence Interval for Global Sales

# Calculate the sample mean and standard deviation
sample_mean = df["Global Sales"].mean()
sample_std = df["Global Sales"].std()

# Determine the sample size
n = len(df)

# Choose a confidence level (e.g., 95%)
confidence_level = 0.95
alpha = 1 - confidence_level

# Find the t-score for the confidence level
t_score = stats.t.ppf(1 - alpha/2, n - 1)

# Calculate the margin of error
margin_of_error = t_score * (sample_std / np.sqrt(n))

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Sample Size: {n}")
print(f"T-Score: {t_score}")
print(f"Margin of Error: {margin_of_error}")
print(f"Confidence Interval: {confidence_interval}")
```

```python
#Hypothesis testing mean
# Hypothesized population mean
mu_0 = 50

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(df["Global Sales"], mu_0)

# Print the test results
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")

# Decide based on p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The average global sales are significantly different from 50.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference from 50.")
```

```python
# Hypothesis testing proportion

from statsmodels.stats.proportion import proportions_ztest

# Calculate the number of PS4 games and total number of games
num_ps4_games = np.sum(df["Platform"] == "PS4")
total_games = len(df)

# Hypothesized proportion
p0 = 0.5

# Perform proportions z-test
count = num_ps4_games
nobs = total_games
stat, p_value = proportions_ztest(count, nobs, p0)

# Print the test results
print(f"Z-Statistic: {stat}")
print(f"P-Value: {p_value}")

# Decide based on p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportion of PS4 games is significantly different from 50%.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference from 50%.")
```

```python
# Taking varoius tests to find relationships
from scipy.stats import ttest_ind, chi2_contingency

# T-test: Compare sales between two platforms (e.g., PC and PS4)
pc_sales = df[df['Platform'] == 'PC']['Global Profit']
ps4_sales = df[df['Platform'] == 'PS4']['Global Profit']
t_stat, p_value = ttest_ind(pc_sales, ps4_sales)
print(f"T-test: t-statistic = {t_stat}, p-value = {p_value}")

# Chi-Square Test: Relationship between genre and platform
contingency_table = pd.crosstab(df['Genre'], df['Platform'])
chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f"Chi-Square Test: chi2 = {chi2}, p-value = {p}")
```

```python
from scipy.stats import pearsonr

# Correlation between Year and Global Sales
correlation, _ = pearsonr(df['Year'].dt.year, df['Global Sales'])
print(f'Correlation between Year and Global Sales: {correlation:.2f}')

# Grouping by year to analyze trends
yearly_sales = df.groupby(df['Year'].dt.year)['Global Sales'].sum()
print(yearly_sales)
```

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA

# Aggregate sales by year
ts = df.groupby(df['Year'].dt.year)['Global Sales'].sum()

# Check for stationarity
result = adfuller(ts)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

# Differencing to make the series stationary if necessary
ts_diff = ts.diff().dropna()

# ARIMA model
model = ARIMA(ts_diff, order=(5, 1, 0))
model_fit = model.fit()
print(model_fit.summary())

# Forecasting
forecast = model_fit.forecast(steps=12)

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(ts, label='Historical Sales',marker='o')
future_years = [ts.index.max() + i for i in range(1, 13)]
plt.plot(future_years, forecast.cumsum() + ts.iloc[-1], label='Forecasted Sales', color='green',marker= 'o')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Video Game Sales Forecast')
plt.legend()
plt.show()
```
